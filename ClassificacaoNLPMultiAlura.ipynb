```python
lista_de_tags = list()
for tags in perguntas.Tags.unique():
  for tag in tags.split():
    if tag not in lista_de_tags:
      lista_de_tags.append(tag)
print(lista_de_tags)
```

```python
def nova_coluna(lista_tag, dataframe, nome_tags):
  for tag in lista_tag:
    coluna = list()
    for linha_tag in dataframe[nome_tags]:
      if tag in linha_tag:
        coluna.append(1)
      else:
        coluna.append(0)
    dataframe[tag] = coluna
  return dataframe
```

```python
nova_coluna(lista_de_tags, perguntas, 'Tags')
```

```python
lista_zip_tags = list(zip(perguntas[lista_de_tags[0]],
                     perguntas[lista_de_tags[1]],
                     perguntas[lista_de_tags[2]],
                     perguntas[lista_de_tags[3]]))
perguntas['todas_tags'] = lista_zip_tags
perguntas
```

```python
from sklearn.model_selection import train_test_split

perguntas_treino, perguntas_teste, tags_treino, tags_teste = train_test_split(perguntas.Perguntas, perguntas.todas_tags, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vetorizar = TfidfVectorizer(max_features=5000, max_df=0.85)
vetorizar
```

```python
vetorizar.fit(perguntas.Perguntas)
perguntas_treino_tfidf = vetorizar.transform(perguntas_treino)
perguntas_teste_tfidf = vetorizar.transform(perguntas_teste)
print(perguntas_treino_tfidf.shape)
print(perguntas_teste_tfidf.shape)
```

```python
import numpy as np

tags_treino_array = np.asarray(list(tags_treino))
tags_teste_array = np.asarray(list(tags_teste))
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression

regressao_logistica = LogisticRegression(solver = 'lbfgs')
classificador_onevsrest = OneVsRestClassifier(regressao_logistica)
classificador_onevsrest.fit(perguntas_treino_tfidf, tags_treino_array)
resultado_onevsrest = classificador_onevsrest.score(perguntas_teste_tfidf, tags_teste_array)
print(resultado_onevsrest)
```

```python
from sklearn.metrics import hamming_loss

previsao_onevsrest = classificador_onevsrest.predict(perguntas_teste_tfidf)
hamming_loss_onevsrest = hamming_loss(tags_teste_array, previsao_onevsrest)
print(hamming_loss_onevsrest)
```

```python
from skmultilearn.problem_transform import ClassifierChain

classificador_chain = ClassifierChain(regressao_logistica)
classificador_chain.fit(perguntas_treino_tfidf, tags_treino_array)
resultado_chain = classificador_chain.score(perguntas_teste_tfidf, tags_teste_array)
print(resultado_chain)
previsao_chain = classificador_chain.predict(perguntas_teste_tfidf)
hamming_loss_chain = hamming_loss(tags_teste_array, previsao_chain)
print(hamming_loss_chain)
```

```python
from skmultilearn.problem_transform import BinaryRelevance

classificador_binary = BinaryRelevance(regressao_logistica)
classificador_binary.fit(perguntas_treino_tfidf, tags_treino_array)
resultado_binary = classificador_binary.score(perguntas_teste_tfidf, tags_teste_array)
previsao_binary = classificador_binary.predict(perguntas_teste_tfidf)
hamming_loss_binary = hamming_loss(tags_teste_array, previsao_binary)
print(hamming_loss_binary)
print(resultado_binary)
```
